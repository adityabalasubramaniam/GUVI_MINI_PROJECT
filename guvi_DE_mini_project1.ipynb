Proceding to the next Task 4 - find and processing Missing data

```python
missing_percentage_before = df.isnull().mean() * 100
print("Percentage of missing data before filling:")
print(missing_percentage_before)
missing_percentage_before.to_list()
```

```python
pd.reset_option('display.max_columns')
```

```python
df['Population'] = df['Male'] + df['Female']
df['Literate'] = df['Literate_Male'] + df['Literate_Female']

# Filling missing values using .loc to avoid chained assignment warnings
df.loc[df['Population'].isnull(), 'Population'] = (
    df['Young_and_Adult'] + df['Middle_Aged'] + df['Senior_Citizen'] + df['Age_Not_Stated']
)

df.loc[df['Households'].isnull(), 'Households'] = (
    df['Households_Rural'] + df['Households_Urban']
)
```

```python
df
```

```python
# Step 4: Calculate percentage of missing data for each column after filling
missing_percentage_after = df.isnull().mean() * 100
print("\nPercentage of missing data after filling:")
print(missing_percentage_after)
```

```python
# Step 5: Compare the two percentages
comparison = pd.DataFrame({
    'Before Filling': missing_percentage_before,
    'After Filling': missing_percentage_after
})

print("\nComparison of missing data percentages:")
print(comparison)
```
